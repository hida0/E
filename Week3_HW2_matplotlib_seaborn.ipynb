### **matplotlib** 라이브러리로 그래프 그리기

1. **전체 그래프가 위치할 기본 틀**을 만든다. plt의 **figure** 메서드를 이용한다.
2. 그래프를 그려 넣을 **그래프 격자**를 만든다. 기본 틀에 add_subplot 메서드를 이용하여 그래프 격자를 그린다.

   **add_subplot(그래프 기본 틀의 행 크기, 그래프 기본 틀의 열 크기, 그래프를 추가할 위치 순서)**
3. 만든 격자에 **plot** 메서드를 이용하여 그래프를 하나씩 추가한다. 격자에 그래프가 추가되는 순서는 **왼쪽에서 오른쪽** 방향이다.
4. 만약 격자의 첫번째 행이 꽉 차면 두번째 행에 그래프를 그려 넣는다. 그래프를 확인하려면 반드시 처음 만든 기본 틀을 입력해야한다.
5. 만든 격자에 **set_title** 메서드를 이용하여 각각의 그래프를 쉽게 구분할 수 있도록 그래프 격자에 제목을 추가한다.
6. **suptitle** 메서드를 이용하여 전체 기본 틀에도 제목을 추가한다.
7. 그래프의 이름과 숫자가 겹쳐 보이는 경우가 생기면 **tight_layout** 메서드를 이용하여 각 그래프의 레이아웃을 보기 좋게 조절한다.

4개의 데이터 그룹으로 구성된 앤스콤 데이터를 모든 데이터 그룹에 대하여 그래프를 그려본다.

```python
dataset_1 = anscombe[anscombe['dataset']=='I']
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']
```

#### 참고 [matplotlib 사용법 기초](https://engineer-mole.tistory.com/215)
####1. **plt.figure()**
* plt.figure()이 맨 처음에 나오는 경우가 많다. figure()은 Figure 인스턴스를 생성하는데 Figure 인스턴스의 역할은 이미지 전체의 영역을 확보하는 것이다. plt.figure()에서는 이미지 영역의 확보뿐이므로, 그래프에는 아무것도 그려지지 않는다.
* 인수로 아래와 같은 항목을 지정할 수 있다.
figsize : (width, height)의 튜플을 전달한다. 단위는 인치이다. /
dpi : 1인치당의 도트 수 / facecolor : 배경색 / edgecolor : 외곽선의 색

#### 2. **fig.add_subplot()**
* plt.figure()으로 Figure 인스턴스를 그리고 그 위에 그래프를 그리기 위해서는 subplot를 추가할 필요가 있다. subplot의 추가는 add_subplot 메소드를 사용한다.
* add_subplot(111)의 의미는 1행째의 1열의 첫 번째라는 의미로, add_subplot(1, 1, 1) 로 작성해도 동일하게 작동한다. 
* add_subplot는 Axes 오브젝트를 반환한다. Axes는 그래프의 이미지, 축의 메모리, 라벨의 설정 등을 맡고 있다.
* add_subplot는 기본적으로 덮어쓴다.

#### 3. **plt.subplot()**
* plt.subplot()는 add_subplot와 동일하게 인수에 행의 수, 열의 수 및 몇 번째 등을 지정할 수 있다. 
* add_subplot과 다른 점은 plt.figure로 만든 Figure 인스턴스가 아닌, 현재의 이미지영역(fig=figure())에 추가하는 메소드라는 점이다.
* subplot는 subplot이전에 그려진 figure과 겹치는 경우, 이전의 figure을 지우는 성질을 가지고 있다.

#### 4. **plt.subplots()**
* plt.subplot()와 비슷한 것으로는 plt.subplots()가 있다. **반환 값은 fig와 Axes 혹은 Axes 객체의 배열**이다. subplots()도 add_subplot의 경우와 동일하게 Axes 객체를 반환한다.
* plt.subplots()는 실제로 fig = figure()를 한 후에 fig.add_subplot(111) 한 것과 동일하다. 
* plt.subplots()의 인수로 행의 수, 열의 수를 전달하는 것으로 여러 개의 Axes 객체를 생성할 수 있다.

#### 5. **Axes**
* Axes 객체는 실제 이미지 데이터의 역할을 한다. 
* Axes 객체에 대해 그려질 데이터를 전달하거나, set_xlabel, set_ylabel, set_title로 라벨이나 타이틀의 설정을 할 수 있다. 또한, 동일하게 Axes 객체에 plot를 겹칠 수도 있다.

#### 6. **plt.title(), plt.xlabel(), plt.ylabel()**
* 그래프 하나씩 그릴 때에는 plt.title(), plt.xlabel(), plt.ylabel()를 자주 사용한다.
* 결국은 ax = subplot(111)하여, ax.set_title(), ax.set_xlabel(), ax.set_ylabel()하는 것과 동일하다.

seaborn 라이브러리에서 tips 데이터 집합을 불러와 다양한 그래프를 그려본다.

#### **1. 히스토그램**
* 데이터프레임의 열 데이터 분포와 빈도를 살펴보는 용도로 자주 사용하는 그래프
* 변수를 하나만 사용해서 그리는 "일변량 그래프"
* **hist** 메서드에 원하는 변수를 전달하며 **bins** 인자값을 지정하여 x축의 간격을 조정할 수 있다.

#### **2. 산점도 그래프**
* 변수 2개를 사용해서 만드는 "이변량 그래프"

```python
tips = sns.load_dataset("tips")
print(type(tips))
tips.head()
```

```python
# 히스토그램
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title("Histogram of Total Bill") # 그래프 제목 추가
axes1.set_xlabel("Frequency") # x축 제목 추가
axes1.set_ylabel("Total Bill") # y축 제목 추가
plt.show()
```

```python
# 산점도 그래프
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title("Scatterplot of Total Bill vs Tip")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")
plt.show()
```

```python
# 박스 그래프
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot([ tips[tips['sex']=='Female']['tip'],
                tips[tips['sex']=='Male']['tip'] ],
                labels=['Female', 'Male'])
axes1.set_xlabel("Sex")
axes1.set_ylabel("Tip")
axes1.set_title("Boxplot of Tips by Sex")
plt.show()
```

### 다변량 그래프 그리기
* 3개 이상의 변수를 사용한 그래프
* 산점도 그래프에 점의 색상을 다르게 하거나 점의 크기를 다르게 하는 등 3개 이상의 변수를 사용하는 다변량 그래프에서는 적재적소에 맞는 그래프 요소를 추가하여 그래프를 표현해야 한다.
* **scatter** 메서드의 **c 인자는 점의 색상(color)**을, **s 인자는 점의 크기(size)**를 의미한다.
* **alpha** 인잣값을 지정하여 점의 투명도를 조절할 수 있다.

```python
# 문자열은 산점도 그래프의 색상을 지정하는 값으로 사용할 수 없으므로 정수로 치환
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1

tips['sex_color'] = tips['sex'].apply(recode_sex)
tips.head(3)
```

```python
# 산점도 그래프 - 앞에서 실습한 그래프에 성별(sex)과 테이블당 인원수(size)를 새로운 변수로 추가하기
# 성별은 점의 색상으로, 테이블당 인원수는 점의 크기로 표현
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'], 
    y = tips['tip'],
    c = tips['sex_color'], alpha=0.5,
    s = tips['size']*10,
    )
axes1.set_title("Total Bill vs Tip Colored by Sex and Sized by Size")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")
plt.show()
```

### 다양한 종류의 **이변량** 그래프 그리기
#### 먼저 **matplotlib(plt)의 subplots** 메서드로 기본 틀을 만든다.
#### **1. 산점도 그래프**
* **regplot** 메서드를 사용 x축 인자와 y축 인자를 입력하여 산점도 그래프와 회귀선을 함께 그린다.
* 회귀선을 제거하려면 **fit_reg 인자를 False**로 지정한다.
* **lmplot** 메서드도 사용 가능하다.

#### **2. jointplot : 산점도 그래프, 히스토그램, 육각 그래프**
* **jointplot** 메서드를 이용하면 **산점도 그래프와 히스토그램**을 한번에 그려준다.
* x와 y 인자에 원하는 열의 이름을 지정하여 그래프를 그린다.
* `set_axis_labels(xlabel=x축 이름, ylabel=y축 이름)`을 이용하여 x축, y축의 이름을, `fig.suptitle(제목 이름)`을 이용하여 그래프 이름을 추가할 수 있다.
* **kind 인자를 hex로 지정하여 육각 그래프(hexbin)**를 그려서 그래프가 겹쳐보일 때 데이터를 쉽게 구분할 수 있다. 
* 육각 그래프는 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프이다.

#### **3. 이차원 밀집도**
* **kdeplot** 메서드를 이용하여 이차원 밀집도를 그릴 수 있다.
* 두개의 열 데이터를 전달한다.
* **shade 인자값을 True**로 지정하면 그래프에 음영 효과를 줄 수 있다.



```python
# 산점도 그래프
ax = plt.subplots() # matplotlib(plt)로 기본 틀 제작
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title("Scatterplot of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
plt.show()
```

```python
# 회귀선을 제외한 산점도 그래프
ax = plt.subplots() # matplotlib(plt)로 기본 틀 제작
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title("Scatterplot of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
plt.show()
```

```python
# 산점도 그래프와 히스토그램
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle("Joint Plot of Total Bill and Tip", fontsize=10, y=1.03)
plt.show()
```

```python
# 육각 그래프
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle("Hexbin Joint Plot of Total Bill and Tip", fontsize=10, y=1.03)
plt.show()
```

## [seaborn 라이브러리로 만든 그래프 스타일 설정하기]
* **set_style** 메서드를 사용하여 seaborn 라이브러리로 만든 그래프의 스타일을 변경할 수 있다. 
* set_style은 스타일을 적용할 부분에서 한번만 실행하면 되고, 그 이후의 그래프는 모두 지정한 스타일로 바뀐다.
* 그래프 스타일의 종류에는 whitegrid, darkgrid, dark, white, ticks가 있다.

```python
# 그래프 스타일 변경 이전
sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
plt.show()
```

```python
# 그래프를 whitegrid 스타일로 설정
sns.set_style('whitegrid')
sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
plt.show()
```

```python
# for문을 이용하여 모든 스타일을 하나씩 적용
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles): # (0, 'darkgrid')...의 tuple 생성
    plot_position = idx+1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position) # 2행 3열의 그리드에 1번부터 5번까지 그래프를 그린다
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
