## **1. datetime 오브젝트**
* 날짜와 시간을 처리하는 등의 다양한 기능을 제공하는 파이썬 라이브러리다. 
* 날짜를 처리하는 **date** 오브젝트, 시간을 처리하는 **time** 오브젝트, 날짜와 시간을 모두 처리하는 **datetime** 오브젝트가 포함되어 있다.
* **datetime**으로부터 **datetime**을 임포트하여 사용한다.
* **`now`**, **`today`** 메서드를 사용하면 현재 시간을 출력할 수 있다.
* datetime의 인자로 입력한 시간을 바탕으로 datetime 오브젝트를 생성할 수 있다.
* 여러 개의 datetime 오브젝트를 이용하면 datetime 오브젝트 간의 시간 차이를 계산할 수 있다.


```python
from datetime import datetime
```

```python
# now 메서드 사용
now = datetime.now()
print(now)

# today 메서드 사용
today = datetime.today()
print(today)
```

```python
# 원하는 시간을 출력
t1 = datetime.now()
t2 = datetime(1970, 1, 1)
t3 = datetime(1970, 12, 12, 13, 24, 34)

print(t1)
print(t2)
print(t3)
```

```python
# datetime 오브젝트 간의 차이
diff1 = t1-t2
print(diff1)

diff2 = t2-t1
print(diff2)
```

### 테슬라 주식 데이터로 시간 계산하기
#### **3. 시간 범위와 인덱스**
* 임의의 시간 범위를 생성하여 인덱스로 지정한다.
* 특정 일에 데이터가 누락되었을 때, 누락된 데이터도 포함시켜 데이터를 살펴볼 때 사용할 수 있다.
* pandas의 **`date_range`** 메서드의 인자로 **start**와 **end**에 날짜를 입력하여 원하는 시간 인덱스를 생성할 수 있다.
* 새롭게 만든 시간 범위를 인덱스로 사용하기 위해서는 반드시 'Date' 열을 먼저 인덱스로 지정한 후, 생성한 시간 범위를 인덱스로 지정해야 한다.



> **`date_range`** 메서드의 **freq** 인자로 사용할 수 있는 시간 주기의 대표적 예시
* B : 평일만 포함
* W : 주간 단위
* M : 월 마지막 날만 포함
* MS : 월 시작일만 포함
* Q : 3,6,9,12월 분기 마지막 날만 포함
* QS : 3,6,9,12월 분기 시작일만 포함
* A : 년의 마지막 날만 포함
* AS : 년의 시작일만 포함
* H : 시간 단위로 포함(00:00~00:00)
* T : 분 단위 포함
* S : 초 단위 포함

```python
ebola5 = ebola.head()
ebola5.index = ebola5['Date'] # 'Date'열을 먼저 인덱스로 지정
ebola5.reindex(head_range) # 새로 생성한 시간 범위를 인덱스로 지정
ebola5.iloc[:,:5]
```

```python
# date_range의 freq 인자 사용
pd.date_range('2017-01-01', '2017-01-07', freq='B') # B는 평일만 포함하므로 일요일인 1일과 토요일인 7일은 포함되지 않음
```
